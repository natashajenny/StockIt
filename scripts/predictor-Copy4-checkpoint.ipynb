```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
df = pd.read_parquet('abc.parquet')
dataset = df.values
```

```python
# Variables
BATCH_SIZE = 30
EPOCHS = 10
DROPOUT = 0.25
LSTM_UNITS = 100
TRAIN_LIMIT = 0.8
```

```python
# Data split
cut_off = int(np.floor(TRAIN_LIMIT * dataset.shape[0]))
train = dataset[:cut_off, :]
valid = dataset[cut_off - BATCH_SIZE:, :]
```

```python
# Scale Training
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train)
valid_scaled = scaler.transform(valid)
```

```python
# Batch up data
x_train, y_train = [], []
for i in range(BATCH_SIZE, len(train)):
    x_train.append(train_scaled[i - BATCH_SIZE:i, :-1])
    y_train.append(train_scaled[i, -1])
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Define model
model = Sequential()
model.add(LSTM(units=LSTM_UNITS, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(Dropout(DROPOUT))
model.add(LSTM(units=LSTM_UNITS))
model.add(Dropout(DROPOUT))
model.add(Dense(256))
model.add(Dense(32))
model.add(Dense(1))
```

```python
# Train model
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1)
```
